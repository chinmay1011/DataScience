```python
import pandas as pd
import numpy as np
```

```python
election_result = pd.read_csv('./data/2016-us-election/primary_results.csv')
```

```python
election_result.head()
```

```python
election_result.shape
```

```python
election_result_state = election_result
del election_result_state['fips']
del election_result_state['fraction_votes']
```

```python
election_result_state.columns
```

```python
election_result_state = election_result_state[['state','state_abbreviation','county','party','candidate','votes']].groupby(['candidate','state'],as_index=False).sum()
```

```python
election_result_state.head(20)
```

```python
election_result_state[['state','candidate','votes']].groupby(['state','candidate'],as_index=False).max()
```

```python
trump_data = election_result_state[election_result_state['candidate']=='Donald Trump']
```

```python
trump_data.head()
```

```python
election_result_state['state'].unique()
```

```python
final_data = pd.DataFrame(election_result_state['state'].unique(), columns=['state'])
```

```python
final_data = pd.merge(final_data, trump_data, on='state', how='left' )
```

```python
final_data.head()
```

```python
final_data['Trump_votes'] = final_data['votes']
del final_data['candidate']
del final_data['votes']
```

```python
final_data.head()
```

```python
clinton_data = election_result_state[election_result_state['candidate']=='Hillary Clinton']
```

```python
clinton_data.head()
```

```python
final_data = pd.merge(final_data, clinton_data, on='state', how='left' )
```

```python
final_data.head()
```

```python
final_data['Clinton_votes'] = final_data['votes']
del final_data['candidate']
del final_data['votes']
```

```python
final_data.head()
```

```python
total_data = election_result_state
```

```python
total_data.head()
```

```python
total_data['candidate'].unique()
```

```python
total_data = total_data[['state','votes']].groupby('state',as_index=False).sum()
```

```python
total_data.head()
```

```python
total_data.isnull().any()
```

```python
total_data.shape
```

```python
final_data.shape
```

```python
final_data = pd.merge(final_data, total_data, on='state', how='inner')
```

```python
final_data.head()
```

```python
final_data['Total_votes'] = final_data['votes']
```

```python
final_data.describe().transpose()
```

```python
final_data['Trump_%'] = (final_data['Trump_votes']/final_data['votes'])*100
```

```python
final_data.head()
```

```python
final_data['Clinton_%'] = (final_data['Clinton_votes']/final_data['votes'])*100
```

```python
final_data.head()
```

```python
len(final_data[final_data['Trump_%']<final_data['Clinton_%']])
```
